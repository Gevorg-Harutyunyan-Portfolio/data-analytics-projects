## Import Section

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib import cm
import matplotlib.dates as mdates
from matplotlib.colors import Normalize
import calendar
from scipy import stats
import random
from matplotlib.colors import to_hex
```

```python
# Load Data
calendar = pd.read_csv("calendar.csv")
listings = pd.read_csv("listings.csv")
```

## Data Cleaning

### Cleaning calendar data

```python
calendar["price"] = calendar["price"].str.lstrip("$").str.replace(",", "")
calendar["price"] = calendar["price"].astype(float)
```

```python
# Convert date to datetime
calendar['date'] = pd.to_datetime(calendar['date'])
```

```python
# Convert 'available' to boolean
calendar["available"] = calendar["available"].replace("t",True).replace("f",False)
```

```python
# Drop rows with NaN listing_id or date
calendar.dropna(subset=['listing_id', 'date'], inplace=True)
```

### Cleaning listings data

```python
listings["price"] = listings["price"].str.lstrip("$").str.replace(",", "")
listings["price"] = listings["price"].astype(float)
```

```python
# Convert booleans from t/f to True/False
bool_columns = ['instant_bookable', 'host_identity_verified', 'host_has_profile_pic', 'host_is_superhost']
for col in bool_columns:
    listings[col] = listings[col].replace("t",True).replace("f",False)
```

```python
# Convert datetime columns
date_cols = ['host_since', 'last_scraped', 'calendar_last_scraped', 'first_review', 'last_review']
for col in date_cols:
    listings[col] = pd.to_datetime(listings[col], errors='coerce')
```

```python
# Convert numerical columns
num_cols = ["price",'minimum_nights', 'maximum_nights', 'number_of_reviews', 'review_scores_rating']
for col in num_cols:
    listings[col] = pd.to_numeric(listings[col], errors='coerce')
```

```python
# Columns whish will be used
cols_to_use = ['id', 'name', 'neighbourhood_cleansed', 'room_type', 'price', 'minimum_nights', 'maximum_nights',
                'availability_365', 'number_of_reviews', 'review_scores_rating', 'host_is_superhost','host_listings_count',
                'host_total_listings_count', 'host_identity_verified', 'bedrooms',
                'latitude', 'longitude', 'instant_bookable']
listings_clean = listings[cols_to_use]
```

### Aggregating calendar data


```python
# Prices sre stable for each listing_id and min, mean, max are equal.
# That's why thsi part I haven't used further

calendar_grouped = calendar.groupby('listing_id').agg({
    'price': ['mean', 'median', 'min', 'max', 'std'],
    'available': 'sum'
}).reset_index()
calendar_grouped.columns = ['listing_id', 'mean_price', 'median_price', 'min_price', 'max_price', 'std_price', 'days_available']
```

### Merging listings and calendar data

```python
merged_df = listings_clean.merge(calendar_grouped, left_on='id', right_on='listing_id')
merged_df.drop(columns=['listing_id'], inplace=True)
```

### Some results

```python
print("Basic Info:")
print(merged_df.info())
```

## Visualization

#### Plot 1. Boxplot: Price Distribution by Neighborhood

```python
merged_df = merged_df.dropna(subset=['neighbourhood_cleansed', 'price'])
grouped_data = merged_df.groupby('neighbourhood_cleansed')['price'].apply(list)
price_data = grouped_data.tolist() 
neighborhoods = grouped_data.index.tolist()
plt.figure(figsize=(12, 9))
boxplot_result = plt.boxplot(price_data, showfliers=True, patch_artist=True)
for box in boxplot_result['boxes']:
    box.set_facecolor('lightblue')
for median in boxplot_result['medians']:
    median.set_color('red')
plt.xticks(ticks=range(1, len(neighborhoods) + 1), labels=neighborhoods, rotation=90)
plt.title('Price Distribution by Neighborhood', fontsize=16)
plt.ylabel('Price (€)', fontsize=15)
plt.tight_layout()
```

#### 2. Scatter Plot: Price vs Number of Reviews

```python
plt.figure(figsize=(10,6))
plt.scatter(merged_df['number_of_reviews'], merged_df['price'], alpha=0.5, color='dodgerblue')
plt.title('Price vs Number of Reviews', fontsize=13)
plt.xlabel('Number of Reviews', fontsize=12)
plt.ylabel('Price (€)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.2)
plt.tight_layout()
plt.show()
```

#### 3. Histogram: Distribution of Bedrooms

```python
plt.figure(figsize=(10, 6))
plt.hist(merged_df['bedrooms'].dropna(), bins=7, color='orange', edgecolor='black')
plt.title('Distribution of Bedrooms')
plt.xlabel('Bedrooms', fontsize=13)
plt.ylabel('Frequency', fontsize=13)
plt.grid(True)
plt.show()
```

#### 5. Boxplot: Bedrooms vs Price

```python
bedroom_groups = [merged_df[merged_df['bedrooms'] == i]['price'] for i in sorted(merged_df['bedrooms'].dropna().unique()) if i <= 5]
plt.figure(figsize=(10,6))
boxplot_result=plt.boxplot(bedroom_groups, showfliers=True, patch_artist=True)
for box in boxplot_result['boxes']:
    box.set_facecolor('dodgerblue')
for median in boxplot_result['medians']:
    median.set_color('yellow')
plt.title('Price Distribution by Bedroom Count (1–5)', fontsize=14)
plt.xlabel('Bedrooms',fontsize=13)
plt.ylabel('Price (€)', fontsize=13)
plt.grid(True, linestyle='--', alpha=0.2)
plt.tight_layout()
plt.show()
```

#### 10. Scatter Plot: Latitude vs Longitude

```python
plt.figure(figsize=(12, 8))

# Get unique neighborhoods
unique_neighborhoods = merged_df['neighbourhood_cleansed'].unique()

num_neighborhoods = len(unique_neighborhoods)
cmap = plt.cm.get_cmap('viridis', num_neighborhoods) # You can change viridis to another cmap
neighborhood_colors = [to_hex(cmap(i)) for i in range(num_neighborhoods)]

for i, neighborhood in enumerate(unique_neighborhoods):
    neighborhood_data = merged_df[merged_df['neighbourhood_cleansed'] == neighborhood]
    color = neighborhood_colors[i]
    
    # Plot neighborhood data with its specific color
    plt.scatter(neighborhood_data['longitude'], neighborhood_data['latitude'], 
                color=color, label=neighborhood, alpha=0.8, s=100, edgecolors='none')

plt.legend(loc='center left', bbox_to_anchor=(1.05, 0.5), 
           ncol=1, fontsize=12.5, frameon=True, 
           title='Neighborhoods', title_fontsize=13)
plt.title('Geographic Distribution of Airbnb Listings', fontsize=16)
plt.xlabel('Longitude', fontsize=14)
plt.ylabel('Latitude', fontsize=14)
```
